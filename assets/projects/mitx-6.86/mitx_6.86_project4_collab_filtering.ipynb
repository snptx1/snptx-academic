```python
import numpy as np
import common
import kmeans

# Load the toy dataset
X = np.loadtxt('toy_data.txt')

# Define the range of K and the seeds
K_values = [1, 2, 3, 4]
seeds = [0, 1, 2, 3, 4]

# Initialize an empty dictionary to store the lowest cost for each K
lowest_costs = {}

# Iterate over the values of K
for K in K_values:
    best_cost = float('inf')
    best_mixture = None
    best_post = None

    # Iterate over the seeds
    for seed in seeds:
        # Initialize the parameters
        mixture, post = common.init(X, K, seed)

        # Run the K-means algorithm
        mixture, post, cost = kmeans.run(X, mixture, post)

        # Check if the current cost is lower than the best cost
        if cost < best_cost:
            best_cost = cost
            best_mixture = mixture
            best_post = post

    # Store the lowest cost for the current K
    lowest_costs[K] = best_cost

    # Plot the best solution for the current K
    common.plot(X, best_mixture, best_post, f'K={K}, Seed={seeds[np.argmin(lowest_costs)]}')

# Print the lowest costs for each K
for K in K_values:
    print(f'Cost|K={K}= {lowest_costs[K]}')
```
